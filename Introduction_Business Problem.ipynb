### Business Problem
Housing and accomodation is always a concern for many people living in cities. People are always looking out for the best possible deals with regards to their living arrangements. So, in this project I will seek to assist people who are looking to rent an apartment in Vienna. This project will be structured using the following outline:
#### If they are looking to move to Vienna they can see:
* which district has cheaper rent. And/or,
* they can choose to live in residential or commercial areas and can see for example which residential district is best

#### Also, if they already live in one of the 23 districts in Vienna they will be able to see:
* If they are paying more than the average price for their apartment. And/or,
* If there are similar districts to theirs with lower rents
