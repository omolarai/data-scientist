# Data
* The data on apartments (size, number of rooms, address, and price) is collected by scraping a local website with apartment listings (willhaben.at)
* Using geopy, I find the coordinates for each district, and then
* using Foursquare I collect the closest venues (supermarket, restaurant, park, etc.
* After the data collection I can run k-means clustering to cluster the districts into residential and commercial areas, and then
* visualize all the data on a single choropleth map.
